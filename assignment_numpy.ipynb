* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
from copy import deepcopy

#7.1 without numpy
X = [[1, 0, 1, 5, 5, 7],
     [2, 0, 2, 5, 5, 7],
     [3, 0, 3, 5, 5, 7],
     [4, 4, 4, 5, 5, 7],
     [4, 4, 4, 5, 5, 7],
     [4, 4, 4, 5, 5, 7]]
new_list = deepcopy(X[:len(X[0])])
result_7_1_py = 1
for i,j in enumerate(new_list):
    if j[i] != 0:
        result_7_1_py *= j[i]
print(result_7_1_py)

#7.1 numpy
X = [[1, 0, 1, 5, 5, 7],
     [2, 0, 2, 5, 5, 7],
     [3, 0, 3, 5, 5, 7],
     [4, 4, 4, 5, 5, 7],
     [4, 4, 4, 5, 5, 7],
     [4, 4, 4, 5, 5, 7]]

main_diag = np.diag(X)
result_7_1_np = np.prod(main_diag, where=(main_diag != 0))
print(result_7_1_np)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
from scipy.spatial.distance import cdist
from scipy.spatial import distance
from functools import wraps
import time


#7.5 without numpy
A = [[1, 2, 3, 4],[5, 6, 7, 9]]
B = [0.5, 0.1, 0.3, 0.6]

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        st = time.time()
        res = func(*args, **kwargs)
        dt = time.time() - st
        print(f'Результат выполнения функции : {res}')
        print(f'Время выполнения функции {wrapper.__name__}: {dt}')
    return wrapper


@timer
def evkld_dist_python(xp:list, yp:list)->list:
    """Evklid distance by Python"""
    evkld_dist = []
    for row in xp:
        dist = [pow((i - j),2) for i, j in zip(row, yp)]
        evkld_dist.append((sum(dist))**0.5)
    return evkld_dist


#7.5 with scipy
As = np.array([[1, 2, 3, 4],[5, 6, 7, 9]])
Bs = np.array([[0.5, 0.1, 0.3, 0.6]])

@timer
def evkld_dist_scipy(xs:list, ys:list)->list:
    """Evklid distance by Spicy"""
    evklid_dist = cdist(xs, ys, metric='euclidean')
    return evklid_dist


#7.5 with numpy
An = np.array([[1, 2, 3, 4],[5, 6, 7, 9]])
Bn = np.array([[0.5, 0.1, 0.3, 0.6]])

@timer
def evkld_dist_numpy(xn:list, yn:list)->list:
    """Evklid distance by Numpy"""
    differ = xn - yn
    pow_res = np.pow(differ,2)
    evkld_dist = np.sum(pow_res, axis=1)
    return np.sqrt(evkld_dist)

#вызов 3х функций подсчёта Evklid distance
evkld_dist_python(A, B)
evkld_dist_scipy(As, Bs)
evkld_dist_numpy(An, Bn)
```
